```python
Student = {}


n = int(input("Enter the number of students : "))
for i in range(n):
    student_id = input("Enter student id : ")

    name = input("Enter student name : ")

    try:
       age = int(input("Enter student age : "))
    except ValueError as e:
       print("Wrong",e)

    marks = float(input("Enter student marks(0-100) : "))
    if marks>100:
      print("Invalid Marks,Please re-enter between 0 to 100")
      marks = float(input("Enter student marks : "))
    else:
      pass


    Student[student_id] = {
        'name': name.title(),
        'age': age,
        'marks': marks
    }


def s_record(Student):
  print(Student)

s_record(Student)


def maxmarks(Student):
  highest_marks = max(Student,key=lambda x:Student[x]['marks'])
  highest_student = Student[highest_marks]
  print(f"Student with highest marks is {highest_student['name']} with marks {highest_student['marks']}")

maxmarks(Student)

def average_marks():
    avg = sum(data['marks'] for data in Student.values()) / len(Student)
    print(f"Average marks: {avg}")

average_marks()
```
