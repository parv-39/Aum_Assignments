```python
def rev_string(sentence):
  words = sentence.split(' ') #used to split the sentence
  rev_sentence = ' '.join(reversed(words)) #used to join the the words in reverse order (inbuilt reverse)
  return rev_sentence


x = input("Enter the text")
y = lambda x : rev_string(x)
print(y(x))
```

```python
def rev_string(sentence):
    stack = sentence.split() #all the words are adds in stack
    words = [] #empty list
    while stack:
        words.append(stack.pop()) #it will pop the element from stack and then add it to list
    return " ".join(words) #it is used to combine all the list elements to string


sentence = input("Enter the text")
print(rev_string(sentence))
```
